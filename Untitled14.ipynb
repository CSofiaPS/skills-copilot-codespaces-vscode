```python
import numpy as np
import matplotlib.pyplot as plt

# Parmetros
L = 1.0  # longitud de la barra
omega = 1.0  # velocidad angular en rad/s
t = np.linspace(0, 10, 100)  # tiempo de 0 a 10 segundos

# Posicin del rectngulo S
S = L * np.cos(omega * t)

# Graficar la posicin en funcin del tiempo
plt.plot(t, S)
plt.xlabel('Tiempo (s)')
plt.ylabel('Posicin del rectngulo S (m)')
plt.title('Posicin del rectngulo S en funcin del tiempo')
plt.grid(True)
plt.show()
import sympy as sp

# Definir la variable angular
phi = sp.symbols('phi')

# Velocidad mnima de S cuando phi = 90 grados o pi/2 radianes
phi_min = sp.pi / 2
print(f"El ngulo  para el cual la velocidad de S es mnima es: {phi_min.evalf()} radianes")
# Parmetros
L = 1.0  # longitud de la barra
omega = 1.0  # velocidad angular en rad/s

# Definir el ngulo phi
phi = np.linspace(0, 2 * np.pi, 100)

# Aceleracin del rectngulo
a = -L * omega**2 * np.cos(phi)

# Graficar la aceleracin en funcin del ngulo
plt.plot(phi, a)
plt.xlabel('ngulo  (radianes)')
plt.ylabel('Aceleracin del rectngulo S (m/s^2)')
plt.title('Aceleracin del rectngulo S en funcin del ngulo ')
plt.grid(True)
plt.show()

```
