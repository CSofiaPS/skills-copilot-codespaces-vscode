```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
L = 1.0  # longitud de la barra
omega = 1.0  # velocidad angular en rad/s
t = np.linspace(0, 10, 100)  # tiempo de 0 a 10 segundos

# Posición del rectángulo S
S = L * np.cos(omega * t)

# Graficar la posición en función del tiempo
plt.plot(t, S)
plt.xlabel('Tiempo (s)')
plt.ylabel('Posición del rectángulo S (m)')
plt.title('Posición del rectángulo S en función del tiempo')
plt.grid(True)
plt.show()
import sympy as sp

# Definir la variable angular
phi = sp.symbols('phi')

# Velocidad mínima de S cuando phi = 90 grados o pi/2 radianes
phi_min = sp.pi / 2
print(f"El ángulo φ para el cual la velocidad de S es mínima es: {phi_min.evalf()} radianes")
# Parámetros
L = 1.0  # longitud de la barra
omega = 1.0  # velocidad angular en rad/s

# Definir el ángulo phi
phi = np.linspace(0, 2 * np.pi, 100)

# Aceleración del rectángulo
a = -L * omega**2 * np.cos(phi)

# Graficar la aceleración en función del ángulo
plt.plot(phi, a)
plt.xlabel('Ángulo φ (radianes)')
plt.ylabel('Aceleración del rectángulo S (m/s^2)')
plt.title('Aceleración del rectángulo S en función del ángulo φ')
plt.grid(True)
plt.show()

```
